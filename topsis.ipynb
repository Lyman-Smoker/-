```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('./data1.CSV',error_bad_lines=False) #加入参数)
```

```python
df
```

```python
#定义决策人给出的决策权重向量
w=[0.25,0.25,0.25,0.25]
p=0.6
```

```python
# 获取每一列最大值
max_df=df.max()
max_df
```

```python
# 获取每一列最小值
min_df=df.min()
min_df
```

```python
#对工作时间和距离正向化
df['working_hours']=df['working_hours'].apply(lambda x: max_df[1]-x)
df['distance']=df['distance'].apply(lambda x : max_df[2]-x)
df['is_mental_work']=df['is_mental_work'].apply(lambda x:p if x==0 else 1-p)
df
```

```python
#标准化
square=[]
for c in range(df.shape[1]):
    s=0
    for r in range(df.shape[0]):
        s=s+df.iloc[r,c]*df.iloc[r,c]
    square.append(s)
df['salary']=df['salary'].apply(lambda x : x/np.sqrt(square[0]))
df['working_hours']=df['working_hours'].apply(lambda x : x/np.sqrt(square[1]))
df['distance']=df['distance'].apply(lambda x : x/np.sqrt(square[2]))
df['is_mental_work']=df['is_mental_work'].apply(lambda x : x/np.sqrt(square[3]))
```

```python
df
```

```python
# 加权
idx=['salary','working_hours','distance','is_mental_work']
for i in range(len(idx)):
    df[idx[i]]=df[idx[i]].apply(lambda x: x*w[i])
df
```

```python
#获取正负理想解
max_df=df.max()
min_df=df.min()
df['score']=0
```

```python
df
```

```python
# 计算分数
for r in range(df.shape[0]):
    f=0
    d_p=0
    d_n=0
    for c in range(df.shape[1]-1):
        d_p+=(df.iloc[r,c]-max_df[c])**2
        d_n+=(df.iloc[r,c]-min_df[c])**2
    d_n=np.sqrt(d_n)
    d_p=np.sqrt(d_p)
    df.iloc[r,4]=d_n/(d_n+d_p)
```
